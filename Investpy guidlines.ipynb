```python
# Trying out Investpy
import investpy
```

```python
#Getting Stock Information
df = investpy.get_stock_historical_data(stock ='AAPL',
                                        country = 'United States',
                                        from_date = '01/01/2010',
                                        to_date = '01/01/2020')
df
```

```python
#Getting Fund Information Step 1
# Use .search_quotes function to locate fund in investing.com website
search_result = investpy.search_quotes(text='Vanguard 500 Index Fund Admiral Shares ', products=['funds'],
                                       countries=['united states'], n_results=1)
print(search_result)
```

```python
#Getting Fund Information Step 2
#Use .retrieve_historical_data function to pull out the history closing prices of the fund
historical_data = search_result.retrieve_historical_data(from_date='01/01/2019', to_date='01/01/2020')
historical_data
```

```python
# Pull out some useful information from the fund
information = search_result.retrieve_information()
print(information)
```

```python
# Technical Indicators
technical_indicators = search_result.retrieve_technical_indicators(interval="daily")
print(technical_indicators)
```
