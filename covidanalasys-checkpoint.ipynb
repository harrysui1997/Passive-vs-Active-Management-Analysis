```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
file = Path('covid_usa_history.csv')
covid_cases = pd.read_csv(file).dropna()
covid_cases
```

```python
covid_cases = covid_cases.set_index(pd.to_datetime(covid_cases['date'], infer_datetime_format=True))
covid_cases
```

```python
covid_cases = covid_cases.drop(columns=['date', 
                                        'deathIncrease', 
                                        'inIcuCurrently', 
                                        'hospitalizedIncrease', 
                                        'hospitalizedCurrently', 
                                        'hospitalizedCumulative', 
                                        'negative', 
                                        'negativeIncrease', 
                                        'onVentilatorCumulative', 
                                        'onVentilatorCurrently', 
                                        'positiveIncrease', 
                                        'states', 
                                        'totalTestResults', 
                                        'totalTestResultsIncrease'])
covid_cases
```

```python
covid_cases = covid_cases.sort_index(ascending=True)
covid_cases
```

```python
covid_cases.plot()
```

```python
covid_pct_change = covid_cases.pct_change()
covid_pct_change
```

```python
covid_pct_change.dropna()
```

```python
covid_pct_change.plot()
```

```python
covid_correlation = covid_pct_change.corr()
covid_correlation
```

```python
covid_correlation.style.background_gradient(cmap="summer")
```
